```python
from PIL import Image
from matplotlib import pyplot as plt
from mtcnn.mtcnn import MTCNN
import numpy as np

# Load an image
image = Image.open('/content/image_1.jpg')

# Convert the image to a NumPy array
pixels = np.array(image.convert('RGB'))

# Create the MTCNN detector
detector = MTCNN()

# Detect faces in the image
faces = detector.detect_faces(pixels)

# List to store bounding boxes
bounding_boxes = []

# Plot the image with bounding boxes around detected faces
plt.imshow(pixels)
ax = plt.gca()

for face in faces:
    x, y, width, height = face['box']
    rect = plt.Rectangle((x, y), width, height, fill=False, color='red')
    ax.add_patch(rect)

    # Store the bounding box in the list without the serial number
    bounding_boxes.append([x, y, width, height])

# Display the image with bounding boxes
plt.show()

# Print the list of bounding boxes
print("List of Bounding Boxes:")
for bbox in bounding_boxes:
    print(bbox)

```
